### **Introduction**

Visualization is a fundamental component of mathematical exploration. Graphical representations allow us to examine functional behavior, geometric structure, and multidimensional relationships that may not be immediately evident from symbolic expressions alone.

Python provides powerful visualization tools through the Matplotlib library. These tools enable the creation of two-dimensional plots, three-dimensional surfaces, contour maps, and vector field representations.

This notebook introduces visualization techniques commonly used in mathematical analysis and scientific computing.

By the end of this notebook, you will be able to generate and customize mathematical visualizations using Python.

## 6.1 2D Cartesian Plots for Functions


**Structure:**
```python
x = np.linspace(a, b, N)  # Discretize domain
y = f(x)                  # Evaluate codomain
plt.plot(x, y)            # Render line segments
```

Matplotlib renders continuous functions by sampling $N$ discrete points and connecting them with line segments. `np.linspace` ensures uniform spacing. Higher $N$ yields smoother curves for high-frequency functions.

**Try this:**
```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
plt.plot(x, y)

plt.show()
```

### 6.1.2 Multiple Functions on Same Axes


**Structure:**
```python
plt.plot(x, f(x), label='f')
plt.plot(x, g(x), label='g')
plt.legend()
```

### **6.2 3D Surface and Line Plots**

### 6.2.1 Creating 3D Axes (`projection='3d'`)
**Theoretical Representation:**
$$\mathbb{R}^3 \text{ coordinate system } (x, y, z)$$

**Structure:**
```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
```

Matplotlib requires explicit 3D axis instantiation via `projection='3d'`. Returns an `Axes3D` object with methods like `plot_surface`, `plot_wireframe`, and `plot` for 3D curves.

**Try this:**
```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
```


### 6.2.2 Surface Plots

**Structure:**
```python
X, Y = np.meshgrid(x, y)  # Create 2D grid
Z = f(X, Y)               # Evaluate on grid
ax.plot_surface(X, Y, Z)  # Render surface
```

The `np.meshgrid` converts 1D coordinate vectors into 2D matrices representing a rectangular domain. `plot_surface` interpolates colors based on $Z$ values (height). Requires `X`, `Y`, `Z` to have identical shapes.

**Try this:**
```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)

X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax.plot_surface(X, Y, Z, cmap='viridis')
```

### 6.2.3 3D Parametric Curves 

```python
t = np.linspace(0, 10*np.pi, 100)
x = np.cos(t)
y = np.sin(t)
z = t
ax.plot(x, y, z)  # 3D line plot
```

Used for space curves (e.g., helices, knots). Define parameter $t$, compute component vectors, and pass to `ax.plot()`. Distinct from `plot_surface` which requires 2D grids.

**Try this:**


$$x(t)=cos(t)\\ y(t)=sin(t)$$

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
t = np.linspace(0, 4*np.pi, 200)
ax.plot(np.cos(t), np.sin(t), t)

```

## **6.3 Contour and Density Plots**

### 6.3.1 Contour Lines 

```python
plt.contour(X, Y, Z, levels=20)  # 10 level curves
```

Visualizes 3D surfaces in 2D by drawing lines of constant height (level sets). Useful for topography, equipotential lines, or phase portraits. `levels` controls density of curves.

**Try this:**
```python
X, Y = np.meshgrid(np.linspace(-3,3), np.linspace(-3,3))
Z = X**2 - Y**2
plt.contour(X, Y, Z, levels=20)
```

**Structure:**

The `plt.contourf()` fills regions between level curves with color, creating a heat map effect. Always include `plt.colorbar()` to provide the scale mapping colors to $z$-values.

## 6.4.0 Vector Field Visualization

### 6.4.1 Quiver Plots (`plt.quiver()`, 2D Vector Fields)
**Theoretical Representation:**
$$\mathbf{F}(x,y) = \langle P(x,y), Q(x,y) \rangle$$

**Computational Representation:**
```python
U = P(X, Y)  # x-components
V = Q(X, Y)  # y-components
plt.quiver(X, Y, U, V)
```

**Explanation:**
Draws arrows at grid points $(X,Y)$ with direction/magnitude given by $(U,V)$. Essential for visualizing differential equations ($\dot{\mathbf{x}} = \mathbf{F}(\mathbf{x})$), gradients, or fluid flow.

**Example Structure:**
```python
X, Y = np.meshgrid(np.arange(-2, 3), np.arange(-2, 3))
U, V = -Y, X  # Rotation field
plt.quiver(X, Y, U, V)
```

### 6.4.2 Gradient Field Visualization

$$\nabla f = \left\langle \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y} \right\rangle$$

**Example:**
```python
U, V = np.gradient(Z, dx, dy)  # Numerical gradient
plt.quiver(X, Y, U, V)
```

**Explanation:**
Visualizes the direction of steepest ascent. Arrows should be perpendicular to contour lines of $f$. `np.gradient` computes numerical partial derivatives from the mesh grid.

**Try this:**
```python
Z = X**2 + Y**2
U, V = np.gradient(Z, 0.1, 0.1)
plt.contour(X, Y, Z)
plt.quiver(X, Y, U, V, color='r')
```

## 6.5.0 Interactive Simulations with Widgets

### 6.5.1 Parameter Sliders (`ipywidgets.interact`)

$$f(x; a, b) = a \sin(bx) \quad \text{Parameters } a, b \in \mathbb{R}$$

**Example:**
```python
from ipywidgets import interact
@interact(a=(0, 5, 0.1), b=(1, 10, 1))
def plot(a, b):
    # plotting code using a, b
```

`ipywidgets` creates UI controls (sliders, dropdowns) linked to function arguments. Updates plot dynamically upon interaction. Ideal for exploring parameter sensitivity in mathematical models.

**Example Structure:**
```python
from ipywidgets import interact
import numpy as np, matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
@interact(a=(0, 5, 0.5))
def plot_sin(a):
    plt.plot(x, np.sin(a*x))
    plt.ylim(-5, 5)
```
