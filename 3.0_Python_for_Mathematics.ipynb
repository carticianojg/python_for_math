### **3.1 Single Variable Calculus**

### 3.1.1 Basic Function Definition $y = f(x)$, 2 variables total

$$f: \mathbb{R} \to \mathbb{R}, \quad f(x) = x^2 + 2x + 1$$

Symbolic computation requires declaring symbols before constructing expressions: `x = symbols('x')` creates a mathematical variable rather than a numeric placeholder. Defining $f(x)$ symbolically enables exact manipulation—differentiation, integration, simplification—without numerical approximation errors inherent in floating-point arithmetic.

**Example Structure:**
```python
from sympy import symbols, sin, exp
x = symbols('x')
f = x**2 + 2*x + 1  # Represents f(x) = x² + 2x + 1
```

### 3.1.2 Symbolic Differentiation

$$\frac{df}{dx} = \lim_{h \to 0} \frac{f(x+h) - f(x)}{h}$$


**Example:**
$$\frac{df}{dx} = \frac{d}{dx}(x^2 + 2x + 1)$$

The `diff(f, x)` function computes the exact symbolic derivative $\frac{df}{dx}$ using calculus rules (power, product, chain), returning a new symbolic expression. This is essential for deriving analytical gradients in optimization, verifying manual computations, or generating Jacobian entries for numerical solvers.

**Example Structure:**
```python
from sympy import diff
df_dx = diff(f, x)        # First derivative: f'(x)
d2f_dx2 = diff(f, x, 2)   # Second derivative: f''(x)
```

### 3.1.3 Symbolic Integration

$$F(x) = \int f(x) \, dx \quad \text{s.t.} \quad \frac{dF}{dx} = f(x)$$

**Example:**
$$\int (2x + 2)$$

The `integrate(f, x)` function computes the indefinite integral $\int f(x) \, dx$ symbolically, returning an antiderivative plus constant of integration (omitted by convention). Symbolic integration provides closed-form solutions for theoretical analysis, though some functions require numerical quadrature when no elementary antiderivative exists.

**Example Structure:**
```python
from sympy import integrate
F = integrate(f, x)  # Antiderivative F(x) where F'(x) = f(x)
```

### 3.1.4 Definite Integrals

**Fundamental Theorem of Calculus:**

$$\int_a^b f(x) \, dx = F(b) - F(a)$$

**Example:**

$$\int_0^1 (2x + 2) \, dx$$

Specifying integration bounds as a tuple `(x, a, b)` computes the definite integral $\int_a^b f(x) \, dx$, evaluating the antiderivative at endpoints via the Fundamental Theorem of Calculus. This yields exact results for area, probability mass, or expected value calculations when the integrand is symbolically tractable.

**Example Structure:**
```python
from sympy import Rational
area = integrate(f, (x, 0, 1))
float_area = float(area.evalf())  # Numeric approximation if needed
```

Multivariable functions require declaring multiple symbols: `x, y = symbols('x y')` enables construction of expressions $f(x, y)$ representing surfaces, potentials, or joint distributions. ### 3.2.2 Partial Derivatives

$$\frac{\partial f}{\partial x} = \lim_{h \to 0} \frac{f(x+h, y) - f(x, y)}{h}, \quad \frac{\partial f}{\partial y} = \lim_{k \to 0} \frac{f(x, y+k) - f(x, y)}{k}$$

**Example:**
$$\frac{\partial}{\partial x}(x^2 y) = 2xy$$

Partial derivatives isolate the rate of change with respect to one variable while holding others constant: `diff(f, x)` computes $\frac{\partial f}{\partial x}$ symbolically. This is foundational for gradient-based optimization, sensitivity analysis, and deriving Euler-Lagrange equations in variational calculus.

**Example Structure:**
```python
from sympy import diff
df_dx = diff(f, x)  # ∂f/∂x (y held constant)
df_dy = diff(f, y)  # ∂f/∂y (x held constant)
```

### 3.2.3 Gradient Vector

$$\nabla f(x, y) = \begin{bmatrix} \frac{\partial f}{\partial x} \\ \frac{\partial f}{\partial y} \end{bmatrix} \in \mathbb{R}^2$$

**Example:**
$$\nabla (x^2 + y^2) = \begin{bmatrix} 2x \\ 2y \end{bmatrix}$$

The gradient $\nabla f$ aggregates all first-order partial derivatives into a vector indicating the direction of steepest ascent. Symbolic construction via `[diff(f, x), diff(f, y)]` enables exact gradient expressions for implementing custom optimization algorithms or verifying automatic differentiation outputs.

**Example Structure:**
```python
from sympy import Matrix
gradient = Matrix([diff(f, x), diff(f, y)])  # Column vector ∇f
```

### 3.2.4 Multiple Integrals 

**Fubini's Theorem:**
$$\iint_{\mathcal{D}} f(x,y) \, dx \, dy = \int_c^d \left( \int_a^b f(x,y) \, dx \right) dy$$

**Example:**
$$\int_0^1 \int_0^1 1 \, dx \, dy = \int_0^1 [x]_0^1 \, dy = \int_0^1 1 \, dy = 1$$

Nested integration computes volumes, probabilities, or expected values over multidimensional domains: `integrate(f, x, y)` evaluates $\iint f(x,y) \, dx \, dy$ with order of integration matching argument sequence. For rectangular domains, specify bounds as tuples: `integrate(f, (x, a, b), (y, c, d))`.

**Example Structure:**
```python
from sympy import integrate
volume = integrate(f, (x, 0, 1), (y, 0, 2))  # ∫₀²∫₀¹ f(x,y) dx dy
```

## **3.4 Matrix Operations**

### 3.4.1 Addition & Scalar Multiplication

$$(\mathbf{A} + \mathbf{B})_{ij} = a_{ij} + b_{ij}, \quad (c\mathbf{A})_{ij} = c \cdot a_{ij}$$

**Example:**
$$\begin{bmatrix} a & b \end{bmatrix} + \begin{bmatrix} c & d \end{bmatrix} = \begin{bmatrix} a+c & b+d \end{bmatrix}$$

Matrix addition requires identical shapes and performs element-wise summation: `A + B` computes $\mathbf{A} + \mathbf{B}$. Scalar multiplication `c * A` scales every entry by $c \in \mathbb{R}$, preserving matrix dimensions; both operations are vectorized for performance and align with linear algebra axioms.

**Example Structure:**
```python
B = np.array([[0, 1], [2, 3]], dtype=float)
C = 2 * A + B  # Linear combination: C = 2A + B
```

### 4.4.2 Matrix Multiplication

$$(\mathbf{A}\mathbf{B})_{ij} = \sum_{k=1}^n a_{ik} b_{kj}, \quad \mathbf{A} \in \mathbb{R}^{m \times n}, \mathbf{B} \in \mathbb{R}^{n \times p}$$

**Example:**
$$\begin{bmatrix} a & b \end{bmatrix} @ \begin{bmatrix} c \\ d \end{bmatrix} = \begin{bmatrix} ac + bd \end{bmatrix}$$

The `@` operator performs matrix multiplication $\mathbf{A}\mathbf{B}$ following linear algebra rules (row-column dot products), requiring compatible inner dimensions: $(m \times n) @ (n \times p) \to (m \times p)$. Avoid `*` for matrix products—it performs element-wise (Hadamard) multiplication instead.

**Example Structure:**
```python
B = np.array([[1, 0], [0, 1], [1, 1]], dtype=float)  # 3×2
product = A @ B  # Result: 2×2 matrix, (AB)ᵢⱼ = Σₖ aᵢₖbₖⱼ
```

### 3.4.3 Dot Product

$$\mathbf{u} \cdot \mathbf{v} = \sum_{i=1}^n u_i v_i = \|\mathbf{u}\| \|\mathbf{v}\| \cos \theta$$

**Example:**
$$\begin{bmatrix} u_1 & u_2 \end{bmatrix} \cdot \begin{bmatrix} v_1 \\ v_2 \end{bmatrix} = u_1 v_1 + u_2 v_2$$

The dot product measures vector similarity and projection. In NumPy, `np.dot(u, v)` or `u @ v` (for 1D arrays) computes the scalar sum of element-wise products. Distinct from matrix multiplication (returns matrix) and Hadamard product (returns vector); crucial for orthogonality checks ($\mathbf{u} \cdot \mathbf{v} = 0$).

**Example Structure:**
```python
u = np.array([1, 0, 0], dtype=float)
v = np.array([0, 1, 0], dtype=float)
dot_product = u @ v  # Result: 0.0 (orthogonal)
```

### 3.4.4 Element-wise Operations

**Hadamard product:**
$$(\mathbf{A} \odot \mathbf{B})_{ij} = a_{ij} \cdot b_{ij}$$

**Example:**
$$\begin{bmatrix} a & b \\ c & d \end{bmatrix} \odot \begin{bmatrix} e & f \\ g & h \end{bmatrix} = \begin{bmatrix} ae & bf \\ cg & dh \end{bmatrix}$$

Element-wise multiplication `A * B` (Hadamard product) requires identical shapes and multiplies corresponding entries: $(\mathbf{A} \circ \mathbf{B})_{ij} = A_{ij} B_{ij}$. This operation is distinct from matrix multiplication and appears in variance computations, mask applications, or component-wise nonlinearities in neural networks.

**Example Structure:**
```python
B = np.array([[1, 2, 3], [4, 5, 6]], dtype=float)  # Same shape as A
hadamard = A * B  # Element-wise: (A ∘ B)ᵢⱼ = aᵢⱼbᵢⱼ
```

### 3.4.5 Transpose & Inverse

$$\mathbf{A}^\top: (\mathbf{A}^\top)_{ij} = a_{ji}$$

$$\mathbf{A}\mathbf{A}^{-1} = \mathbf{A}^{-1}\mathbf{A} = \mathbf{I}$$

**Example:**
$$\begin{bmatrix} a & b \\ c & d \end{bmatrix}^\top = \begin{bmatrix} a & c \\ b & d \end{bmatrix}$$

The transpose `A.T` swaps rows and columns ($\mathbf{A}^\top$), essential for inner products and symmetry checks. The inverse `np.linalg.inv(A)` computes $\mathbf{A}^{-1}$ satisfying $\mathbf{A}\mathbf{A}^{-1} = \mathbf{I}$, but requires $\mathbf{A}$ to be square and non-singular; use `np.linalg.solve()` instead of explicit inversion for solving linear systems.

**Example Structure:**
```python
A_square = np.array([[4, 7], [2, 6]], dtype=float)
A_inv = np.linalg.inv(A_square)  # A⁻¹ such that AA⁻¹ = I
identity_check = A_square @ A_inv  # Should approximate I
```

### 3.4.6 Determinant & Norm

$$\det(\mathbf{A}) = \sum_{\sigma \in S_n} \text{sgn}(\sigma) \prod_{i=1}^n a_{i,\sigma(i)}, \quad \|\mathbf{A}\|_F = \sqrt{\sum_{i,j} a_{ij}^2}$$

**Example:**
$$\det \begin{bmatrix} a & b \\ c & d \end{bmatrix} = ad - bc$$

The determinant `np.linalg.det(A)` computes $\det(\mathbf{A})$, indicating invertibility ($\det \neq 0$) and volume scaling factor of the linear transformation. The norm `np.linalg.norm(A)` measures matrix magnitude (default: Frobenius norm), useful for conditioning analysis, convergence criteria, or regularization terms.

**Example Structure:**
```python
det_A = np.linalg.det(A_square)  # Scalar: det(A)
norm_A = np.linalg.norm(A_square, ord='fro')  # ‖A‖_F = √(Σ aᵢⱼ²)
```

### **3.5 Systems of Linear Equations**

### 3.5.1 Formulation 

$$\mathbf{A}\mathbf{x} = \mathbf{b} \iff \begin{bmatrix} a_{11} & \cdots & a_{1n} \\ \vdots & \ddots & \vdots \\ a_{m1} & \cdots & a_{mn} \end{bmatrix} \begin{bmatrix} x_1 \\ \vdots \\ x_n \end{bmatrix} = \begin{bmatrix} b_1 \\ \vdots \\ b_m \end{bmatrix}$$

**Example:**

$$\begin{cases} 2x + y = 5 \\ x + 3y = 5 \end{cases} \implies \begin{bmatrix} 2 & 1 \\ 1 & 3 \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix} = \begin{bmatrix} 5 \\ 5 \end{bmatrix}$$

Linear systems express relationships $A\mathbf{x} = \mathbf{b}$ where $\mathbf{A} \in \mathbb{R}^{m \times n}$ is the coefficient matrix, $\mathbf{x} \in \mathbb{R}^n$ unknowns, and $\mathbf{b} \in \mathbb{R}^m$ observations. Proper formulation requires matching dimensions: $m$ equations for $n$ variables; square systems ($m=n$) may have unique solutions if $\mathbf{A}$ is invertible.

**Example Structure:**

$$3x + y = 9$$
$$x + 2y = 8 $$
```python
A = np.array([[3, 1], [1, 2]], dtype=float)  # Coefficients matrix A ∈ ℝ²ˣ²
b = np.array([9, 8], dtype=float)            # Constants vector b ∈ ℝ²
```



**Practice Task:**

### 3.5.2 Direct Solver

$$\mathbf{x} = \mathbf{A}^{-1}\mathbf{b}$$
Note: This is computed via LU decomposition, not explicit inversion

**Example:**
$$\begin{cases} 3x + 1y = 9 \\ 1x + 2y = 8 \end{cases} \iff \underbrace{\begin{bmatrix} 3 & 1 \\ 1 & 2 \end{bmatrix}}_{\mathbf{A}} \underbrace{\begin{bmatrix} x \\ y \end{bmatrix}}_{\mathbf{x}} = \underbrace{\begin{bmatrix} 9 \\ 8 \end{bmatrix}}_{\mathbf{b}}$$
The function `np.linalg.solve(A, b)` computes the exact solution $\mathbf{x}$ to $A\mathbf{x} = \mathbf{b}$ using LU decomposition, avoiding explicit matrix inversion for numerical stability and efficiency. This is the preferred method for square, non-singular systems in scientific computing, returning $\mathbf{x}$ as a 1D array.

**Example Structure:**
```python
x = np.linalg.solve(A, b)  # Solves Ax = b for x
print(f"Solution: x = {x[0]:.3f}, y = {x[1]:.3f}")
```

### **3.6 Summation**
### 3.6.1 Symbolic Summation 

$$S = \sum_{n=1}^{\infty} a_n \quad \text{Converges if } \lim_{N \to \infty} \sum_{n=1}^{N} a_n = L \in \mathbb{R}$$

**Structure:**
```python
from sympy import symbols, Sum, oo, summation
n = symbols('n', integer=True)
series = Sum(1/2**n, (n, 1, oo))  # Symbolic infinite sum
value = summation(1/2**n, (n, 1, oo))  # Evaluates to 1
```

SymPy handles infinite series symbolically using `oo` (infinity). `Sum()` creates an unevaluated expression (useful for display), while `summation()` attempts to compute the closed-form limit. Returns `oo` for divergent series (e.g., Harmonic series $\sum 1/n$).



**Example**
```python
from sympy import symbols, summation, oo

n = symbols('n', integer=True)
f = 1/2**n

series = Sum(f, (n, 1, oo))  # Symbolic infinite sum
value = summation(f, (n, 1, oo))  # Evaluates to 1
```
