Python uses indentation (typically 4 spaces) to define code blocks for functions, loops, and conditionals, replacing brace-delimited scopes in other languages. Consistent indentation is syntactically required; misaligned blocks raise `IndentationError` and disrupt mathematical logic flow.

The `print()` function accepts multiple arguments separated by commas, automatically inserting spaces and converting non-string types via `str()`. For mathematical output, use f-strings (`print(f"x = {x:.3f}")`) to control numeric precision and embed expressions directly.

Escape sequences format string output: `\n` inserts a newline, `\t` a tab, and `\\` a literal backslash. These are useful for structuring tabular mathematical output or aligning columns in console-based reports without external libraries.

**Example Structure:**

### **2.2 Basic Data Types**

### 2.2.1 Integers ($\mathbb{Z}$): `int`
Integers represent exact whole numbers without fractional components, ideal for indexing arrays, counting iterations, or defining discrete parameters in combinatorial problems. Python `int` has arbitrary precision, avoiding overflow errors common in fixed-width integer types in other languages.

**Example Structure:**
```python
n = 100  # sample size
index = 0  # loop counter
factorial = math.factorial(5)  # exact integer result
```




### 2.2.2 Floats ($\mathbb{R}$): `float`, precision limits
Floats approximate real numbers using IEEE 754 double precision (~15–17 significant digits), suitable for continuous measurements, coefficients, or numerical integration results. Be aware of rounding errors: `0.1 + 0.2 != 0.3` exactly; use `math.isclose()` for comparisons in convergence checks.

**Example Structure:**
```python
pi = 3.141592653589793
epsilon = 1e-8  # tolerance threshold
print(math.isclose(0.1 + 0.2, 0.3))  # True
```



### 2.2.3 Booleans ($\mathbb{B}$): `True`, `False`
Booleans represent binary truth values used in conditional logic, filtering datasets, or controlling algorithm flow (e.g., `while not converged:`). In numerical contexts, booleans enable mask-based indexing: `arr[arr > 0]` selects positive elements without explicit loops.

**Example Structure:**
```python
is_converged = False
mask = [True, False, True]  # for selective processing
```



### 2.2.4 Strings: `str`, indexing, slicing
Strings store textual data such as variable labels, file paths, or LaTeX-formatted annotations for plots. Support indexing (`s[0]`) and slicing (`s[1:4]`) for substring extraction, useful for parsing output filenames or constructing dynamic titles.

**Example Structure:**
```python
label = r"$\int_0^\infty e^{-x} dx$"  # raw string for LaTeX
filename = "results_2024.csv"
print(filename[-4:])  # '.csv'
```



### 2.2.5 Type Conversion: `int()`, `float()`, `str()`, `type()`
Explicit type conversion ensures operations behave as intended: `int(3.9)` truncates to `3`, `float("2.5")` parses strings to numbers, and `str(x)` formats values for output. Use `type(x)` for debugging or dynamic dispatch based on input class.

**Example Structure:**
```python
x = "42"
n = int(x)  # string to integer
print(f"Type: {type(n)}, Value: {n}")
```

### **2.3 Basic Mathematical Operations**

### 2.3.1 Arithmetic Operators: `+`, `-`, `*`, `/`, `//`, `%`, `**`
Python supports standard arithmetic: `+`/`-` for addition/subtraction, `*`/`/` for multiplication/division, `//` for floor division, `%` for modulo (residue), and `**` for exponentiation. These operators follow PEMDAS precedence; use parentheses to enforce explicit grouping in complex expressions.

**Example Structure:**
```python
a, b = 17, 5
quotient = a // b  # 3
remainder = a % b   # 2
power = 2 ** 10     # 1024
```



### 2.3.2 Built-in Functions: `abs()`, `pow()`, `round()`, `divmod()`
Built-in math functions provide common operations without imports: `abs(x)` for magnitude, `pow(x, y)` for exponentiation, `round(x, ndigits)` for controlled precision, and `divmod(a, b)` returning quotient-remainder pairs. These are optimized in C and preferable to manual implementations for clarity and speed.

**Example Structure:**
```python
x = -3.14
print(abs(x), round(x), pow(2, 8), divmod(17, 5))
```




### 2.3.3 The `math` Module: `math.sqrt()`, `math.log()`, `math.pi`
The `math` module provides access to transcendental functions (`sin`, `cos`, `exp`, `log`), constants (`pi`, `e`), and utilities (`ceil`, `floor`, `isclose`). Import once via `import math` and prefix calls (`math.sqrt(x)`) to avoid namespace pollution in larger projects.

**Example Structure:**
```python
import math
r = 5
area = math.pi * r**2
print(f"Area: {area:.3f}")
```

### 2.4 **Basic Logical Operations**

### 2.4.1 Comparison Operators: `==`, `!=`, `<`, `>`, `<=`, `>=`
Comparison operators evaluate relationships between values, returning booleans for use in conditionals or filtering. In numerical code, avoid `==` for floats due to precision errors; prefer `math.isclose(a, b, rel_tol=1e-9)` for robust equality checks.

**Example Structure:**
```python
x = 0.3
print(x == 0.3)  # May be False due to floating-point representation
print(math.isclose(x, 0.3))  # True
```



### 2.4.2 Boolean Logic: `and`, `or`, `not`
Boolean operators combine conditions: `and` requires both true, `or` requires at least one true, `not` inverts truth value. Short-circuit evaluation stops early (e.g., `False and expensive_check()` skips the second operand), useful for guarding expensive computations in iterative methods.

**Example Structure:**
```python
if n > 0 and math.isclose(residual, 0, abs_tol=1e-6):
    print("Converged")
```



### 2.4.3 Chained Comparisons: `0 < x < 1`
Python allows chaining comparisons for interval membership: `0 < x < 1` is equivalent to `(0 < x) and (x < 1)` but more readable. This syntax is particularly useful for validating probability values, normalized features, or domain constraints in optimization.

**Example Structure:**
```python
p = 0.45
if 0 <= p <= 1:
    print("Valid probability")
```



### 2.4.4 Truthiness: Non-zero numbers, non-empty containers
In boolean contexts, non-zero numbers, non-empty strings/lists, and non-`None` objects evaluate as `True`; zero, empty containers, and `None` evaluate as `False`. Avoid implicit truthiness checks on arrays: use `if arr.any():` instead of `if arr:` to prevent ambiguous truth value errors.

**Example Structure:**
```python
x = 0.0
if x:  # False
    print("Non-zero")
    
arr = np.array([0, 0])
if arr.any():  # Correct way to check array contents
    print("Has non-zero element")
```

### **2.5 Data Structures (Lists)**

### 2.5.1 List Creation: `[1, 2, 3]`
Lists store ordered, mutable sequences of heterogeneous elements, ideal for collecting iteration results, parameter sweeps, or discrete data points. Use square brackets with comma separation; nested lists can represent matrices, though NumPy arrays are preferred for numerical linear algebra.

**Example Structure:**
```python
primes = [2, 3, 5, 7, 11]
matrix = [[1, 2], [3, 4]]  # list of lists
```




### 2.5.2 Indexing & Slicing: `L[0]`, `L[1:3]`, `L[::-1]`
Lists support zero-based indexing (`L[0]` first element) and slicing (`L[start:stop:step]`) for subsequence extraction. Negative indices count from the end (`L[-1]` last element); slicing creates shallow copies, useful for partitioning datasets without modifying originals.

**Example Structure:**
```python
data = [10, 20, 30, 40, 50]
print(data[1:4])   # [20, 30, 40]
print(data[::-1])  # reversed list
```




### 2.5.3 List Methods: `.append()`, `.remove()`, `.pop()`, `.sort()`
List methods modify sequences in-place: `.append(x)` adds to end, `.remove(x)` deletes first occurrence, `.pop(i)` removes and returns element at index, `.sort()` orders elements. For numerical workflows, prefer pre-allocating arrays or using NumPy for performance-critical operations.

**Example Structure:**
```python
results = []
for i in range(5):
    results.append(i**2)
results.sort(reverse=True)
```

## **2.6 Functions**

### 2.6.1 Definition: `def func_name(args):`
Functions encapsulate reusable logic with the `def` keyword, accepting parameters and returning results via `return`. In mathematical code, functions model mappings $f: \mathbb{R}^n \to \mathbb{R}^m$, enabling modular implementation of algorithms like root-finding or numerical integration.

**Example:**
$f(x) = x^2 + 2x +1$
```python
def f(x):
    return x**2 + 2*x + 1
```




### 2.6.2 Parameters & Arguments: Positional vs. Keyword
Parameters are defined in the function signature; arguments are values passed during calls. Positional arguments match by order, keyword arguments by name (`func(x=2)`), improving readability and allowing flexible invocation with defaults or overrides.

**Example:** 

$f(a,b) = a^b$
```python
def power(base, exponent=2):
    return base ** exponent

print(power(3))        # 9 (default exponent)
print(power(3, 3))     # 27 (positional)
print(power(base=2, exponent=5))  # 32 (keyword)
```




### 2.6.3 Return Statements: `return value`
The `return` statement exits a function and passes a result to the caller; multiple values can be returned as tuples (`return a, b`). Omitting `return` yields `None`; explicit returns improve code clarity and enable chaining or conditional logic in numerical pipelines.

**Example:**

$ax^2 +bx +c = 0$

Find $x_1, x_2$

```python
def quadratic_roots(a, b, c):
    disc = b**2 - 4*a*c
    if disc < 0:
        return None, None
    sqrt_disc = math.sqrt(disc)
    return (-b + sqrt_disc)/(2*a), (-b - sqrt_disc)/(2*a)
```
